```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
x=np.arange(0,10)
y=2*x
plt.plot(x,y)
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.xlim(0,10)
plt.ylim(0,16)
plt.title('My Graph')
plt.show()
```

```python
#MAtplotlib subplots->
import matplotlib.pyplot as plt
import numpy as np
a=np.linspace(0,10,11)
b=a**4
x=np.arange(0,10)
y=x*2
fig,axes=plt.subplots(nrows=1,ncols=2)
axes[0].plot(x,y)
axes[1].plot(a,b)
```

```python
import numpy as np
import matplotlib.pyplot as plt
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
x=np.linspace(0,10,11)
ax.plot(x,x,label='X Vs X')
#lw is for the line width
lines=ax.plot(x,x**2,color='Purple',lw=2,label='X VS X**2',marker='o',markersize='8',markerfacecolor='red',markeredgewidth='4',markeredgecolor='orange')
ax.legend(title="Functions",facecolor='lightgray',loc=(0.5,1.1))
```

```python
import numpy as np
import matplotlib.pyplot as plt

m = np.linspace(0, 10, 11)  # Mass values from 0 to 10 kg
c = 3.0e8  # Speed of light (m/s)

E = m * c**2  # Energy calculation
fig=plt.figure()
axes=fig.add_axes([0,0,1,1])
axes.plot(m,E)

plt.show()
```
